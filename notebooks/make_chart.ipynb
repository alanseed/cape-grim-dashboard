```python
import pymongo 
import datetime
```

```python
# open the connection to the database 
client = pymongo.MongoClient("mongodb://localhost:27017/")
db = client["cg_data"] 
data = db["obs_data"]
db.obs_data.index_information()
# db.obs_data.create_index('Name')
```

```python
obs = "Almos2WS" 
start = datetime.datetime.fromisoformat("2021-01-01T00:00:00") 
end = datetime.datetime.fromisoformat("2021-01-07T00:00:00")
```

```python
myquery = {'Name':obs,'Time':{'$gte':start,"$lte":end}}
#print(db.obs_data.find(myquery).explain()['executionStats'])
```

```python
results = db.obs_data.find(myquery)    
```

```python
data = []
for doc in results:
    rec = {"Time":doc["Time"], "Value":doc["Value"]} 
    data.append(rec)
print(data)  
```
